# PANDAS FUNCTIONS
## LOC : Lable Based Selection
## ILOC : Index Based Selection

```python
# All the records related to the specific recording appears in the output
data.loc[(data.rating == "PG-13")]
```

```python
# Using AND Operation
data.loc[(data.rating == "PG-13") & (data["release year"] == 2011)]
```

```python
# Using OR Operation
data.loc[(data.rating == "PG-13") | (data["release year"] == 2011)]
```

```python
data.loc[data.rating == "TV-14"]
```

```python
data.loc[(data.rating == "TV-14") , ['title','ratingLevel']]
```

```python
# This is the way to display the columns as per choice.
data.loc[(data.rating == "TV-14") & (data["user rating score"]>= 90)]
```

```python
# This is the way by which we can display the result with the help of index 
data.iloc[[2,3,5]]
```

```python
# This is the way by which we can display the row as per choice and also the columns as per choice with the help of slicing method
data.iloc[[2,3,5], 0:3]
```

```python
# This is the way by which we can display the row as per choice with the help of slicing method
data.iloc[0:3]
```

```python
# This is the way by which we can pop the column from the last that you do not want to be print
data.iloc[0:4, :2]
```
