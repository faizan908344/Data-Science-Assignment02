### 2.  ILOC

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("netflix_titles.csv");
data
```

```python
# The Syntax of LOC is: data.loc [<row label>, <column label>]

# Here we get all the rows of the column "Type"
data.loc[:,"type"]
```

```python
# Here we get all the rows of the column "Type and Column" more then one column.
data.loc[:,["type", "country"]]
```

```python
# Here we merge all the similar values in the "Type" Column
data.loc[:,"type"].value_counts()
```

```python
# Here we find the particular information from any block
# Syntax: data.loc[<row label>, <column label>]
data.loc[1,"cast"]
```

```python
length = len(data.loc[1,"cast"].split(", "))
length
```

```python
data.loc[0:8806,"cast"]
```
